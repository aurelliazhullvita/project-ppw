```python
import networkx as nx
import matplotlib.pyplot as plt

# Membuat graph kosong
G = nx.Graph()

# Menambahkan simpul (nodes)
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
G.add_node(6)
G.add_node(7)
G.add_node(8)
G.add_node(9)

# Menambahkan sisi (edges)
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 4)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)
G.add_edge(4, 6)
G.add_edge(5, 6)
G.add_edge(5, 7)
G.add_edge(5, 8)
G.add_edge(6, 7)
G.add_edge(6, 8)
G.add_edge(7, 8)
G.add_edge(7, 9)

# Menggambar graph
nx.draw(G, with_labels=True, node_color='lightblue', font_weight='bold', node_size=700)

# Menampilkan graph
plt.show()

# Menghitung closeness centrality
closeness = nx.closeness_centrality(G)
print("Closeness Centrality untuk tiap node:")
for node, centrality in closeness.items():
    print(f"Node {node}: {centrality:.4f}")
```
