```
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

def kompas(alamat):
    try:
        # Membuka URL dan parsing HTML
        html = urlopen(alamat)
        data = BeautifulSoup(html, 'html.parser')

        # Menemukan semua item berita populer
        items = data.findAll("h2", {"class": "articleTitle"})
        dates = data.findAll("div", {"class": "articlePost-date"})
        kategoriBerita = data.findAll("div", {"class": "articlePost-subtitle"})
        links = data.findAll("a", {"class": "article-link"})

        # Debugging: print jumlah item yang ditemukan
        print(f"Ditemukan {len(items)} item berita populer.")

        if len(items) == 0:
            print("Tidak ada item yang ditemukan, periksa struktur HTML-nya.")
            return

        # Menyusun hasil ke dalam list
        hasil = []
        for idx in range(len(items)):
            berita = {}
            berita['Judul Berita'] = items[idx].get_text().strip()  # Mendapatkan judul berita dan membersihkan spasi
            # Mengambil link dari setiap berita
            link = links[idx]['href']
            # berita['Link Berita'] = link

            # Mengambil isi berita dari masing-masing link
            try:
                article_html = urlopen(link)
                article_data = BeautifulSoup(article_html, 'html.parser')
                content = article_data.findAll("div", {"class": "read__content"})

                if content:
                    isi_berita = ' '.join([p.get_text() for p in content[0].findAll('p')])
                    berita['Isi Berita'] = isi_berita
                else:
                    berita['Isi Berita'] = 'Tidak ada konten'
            except Exception as e:
                print(f"Kesalahan saat mengambil isi berita: {e}")
                berita['Isi Berita'] = 'Error saat mengambil isi'
            berita['Tanggal'] = dates[idx].get_text()
            berita['Kategori Berita'] = kategoriBerita[idx].get_text()


            # Tambahkan ke hasil list
            hasil.append(berita)

        # Membuat DataFrame dari hasil
        df = pd.DataFrame(hasil)

        # Output DataFrame
        return df

    except Exception as e:
        print(f"Terjadi kesalahan: {e}")

# Memanggil fungsi dengan URL Kompas
data = kompas("https://indeks.kompas.com")

data
```
